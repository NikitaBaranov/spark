```
#collect is type of action
rdd.collect()
```

```
#map is a transformation, it won't be executed untill an action is associated
new_rdd = rdd.map(lambda x: x*2) # the execution with data is delayed until action is found
```

```
l = new_rdd.collect()
```

```
sc.parallelize(l).map(lambda x: x+2).collect()
```

```
rdd.count()
```

```
new_rdd = rdd.map(lambda x: 2*x)
new_rdd.cache() #It will persist the RDD in the RAM rather than recomputing it again.
new_rdd.count()
#if the code looks like this, it won't reuse the new_rdd already created. Rather, will recompute
new_rdd.collect()
```
